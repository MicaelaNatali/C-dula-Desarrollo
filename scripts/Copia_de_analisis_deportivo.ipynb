```python
import pandas as pd

datos = { "Equipo": ["Boca Juniors", "River Plate", "Racing", "Independiente"], "Goles": [12, 10, 8, 6], "Partidos_ganados":[5, 4, 3, 2]}
df = pd.DataFrame(datos)
df
```

```python
df.to_csv("resultados_futbol.csv", index=False)
print("Archivo CSV generado correctamente")
```

```python
import matplotlib.pyplot as plt

plt.bar(df ["Equipo"], df ["Partidos_ganados"])

plt.title ("Rendimiento de equipos")
plt.xlabel("Equipos")
plt.ylabel("Partidos ganados")

plt.show()
```
